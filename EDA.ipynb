Since many values like email and seo are not present in the actual data, I decided to merge them with other categories and assign a different value. I also manipulate the strings to bring them into a unified form.

```python
cfn = ['Отсутствует ICM', 'Тип ICM не указан', 'CPM', 'CPA', 'direct', 'organic', 'smm']
```

```python
# new categories
no_icm = "Отсутствует|нет в справочнике"
no_type = "не указан"
cpm = "CPM"
cpa = "CPA"
direct = "direct|Email|seo"
organic = "organic|other"
smm = "smm"
```

```python
# conditions
conditions = [
    (df['channel_final_name'].str.contains(no_icm)),
    (df['channel_final_name'].str.contains(no_type)),
    (df['channel_final_name'].str.contains(cpm)),
    (df['channel_final_name'].str.contains(cpa)),
    (df['channel_final_name'].str.contains(direct)),
    (df['channel_final_name'].str.contains(organic)),
    (df['channel_final_name'].str.contains(smm))
]
```

```python
# apply
df['channel'] = np.select(conditions, cfn, default='other')
```

```python
df['channel'].value_counts()
```

```python
sns.countplot(data=df, x="channel")
plt.show()
```

From the graphs, it is clear that the data are not at all linearly correlated, it is obvious that it is necessary to use models that are robust to outliers and that are capable of working with nonlinear dependencies.

```python
# removing irrelevant columns that are no longer needed

df = df.drop(columns=['channel_final_name', 'uid', 'id', 'first_bet_date'])
```

```python
# creating new feature

df['fbt_turnover'] = df['m_turnover'] / df['first_bet_sum']
```

```python
# checking the target variable to understand what bins the data can be split into

sns.histplot(data=df, x='revenue', bins=10)
plt.show()
```

```python
df['revenue'].describe()
```

```python
# A lot of negative revenue values.
# Designating class zero as those who have negative revenue in 180 days.
# First class - the most common average players from 0 to 10,000.
# Second class - from 10,000 to 50,0000.
# Third class - everything above.
df['revenue_class'] = pd.cut(df['revenue'], bins=[-float('inf'), 0, 10000, 50000, float('inf')], labels=[0, 1, 2, 3])
```

```python
sns.countplot(data=df, x="revenue_class")
plt.show()
```

```python
df['revenue_class'].value_counts()
```

Thus, we have identified the classes on which the regressor will be trained.
